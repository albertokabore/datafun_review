# Step 3: Calculate Measures


```python
# Ensure the relevant columns are numeric
diabetes_data['Percent'] = pd.to_numeric(diabetes_data['Percent'], errors='coerce')
diabetes_data['Lower 95% CL'] = pd.to_numeric(diabetes_data['Lower 95% CL'], errors='coerce')
diabetes_data['Upper 95% CL'] = pd.to_numeric(diabetes_data['Upper 95% CL'], errors='coerce')
diabetes_data['Standard Error'] = pd.to_numeric(diabetes_data['Standard Error'], errors='coerce')
```

```python
# Calculate Total # of patients with diabetes mellitus (DM)
total_patients_with_dm = diabetes_data['Percent'].count()
print(f"Total # of patients with DM: {total_patients_with_dm}")
```

```python
# Calculate Average percent of diabetes across all strata categories
average_percent_diabetes = diabetes_data['Percent'].mean()
print(f"Average Percent of Diabetes: {average_percent_diabetes:.2f}%")
```

```python
# Calculate Confidence intervals (Lower 95% CL and Upper 95% CL) for diabetes percent
average_lower_95_cl = diabetes_data['Lower 95% CL'].mean()
average_upper_95_cl = diabetes_data['Upper 95% CL'].mean()
print(f"Average Lower 95% CL: {average_lower_95_cl:.2f}%")
print(f"Average Upper 95% CL: {average_upper_95_cl:.2f}%")
```

```python
# Group by strata for more detailed measures
numeric_columns = ['Year', 'Percent', 'Lower 95% CL', 'Upper 95% CL', 'Standard Error']
grouped_data = diabetes_data.groupby('Strata Name')[numeric_columns].mean()
```

```python
print("\nGrouped data by 'Strata Name':")
print(grouped_data)
```

Observation:

•	Income Levels: Lower-income strata ($15,000 to $24,999) have a higher average diabetes percent (12.33%) compared to higher-income strata ($50,000 and above) with 6.70%.

•	Age Groups: Diabetes percent increases with age, from 1.53% in the 18-34 years group to 21.43% in the 65+ years group.

•	Race/Ethnicity: African-Americans have the highest average diabetes percent (14.27%), while Asians have a lower average (7.93%).

•	Education Levels: Individuals with less than a high school education have a higher diabetes percent (16.28%) compared to college graduates (6.80%).

•	Gender: Males have a slightly higher average diabetes percent (9.68%) compared to females (8.80%).



# Create Visualizations for the Dashboard

```python

# Distribution of Percent
plt.figure(figsize=(10, 6))
sns.histplot(diabetes_data['Percent'], kde=True)
plt.title('Distribution of Percent')
plt.xlabel('Percent')
plt.ylabel('Frequency')
plt.show()
```

Observations:

•	The distribution of diabetes percentages is approximately normal, centered around 10%.

•	The most frequent diabetes percentage range is around 9-11%, with a peak frequency of about 35 occurrences.

•	There is a slight right skew in the distribution, indicating a small number of higher diabetes percentages.

•	Lower diabetes percentages (below 5%) and higher percentages (above 15%) are less common.

•	The range of diabetes percentages extends from approximately 1% to 24%.

•	The distribution suggests that most of the population falls within the mid-range of diabetes percentages, with fewer individuals at the extreme ends.



```python
# Box plot of Percent by Strata Name
plt.figure(figsize=(14, 8))
sns.boxplot(x='Strata Name', y='Percent', data=diabetes_data)
plt.title('Box plot of Percent by Strata Name')
plt.xticks(rotation=90)
plt.show()
```

Observations:

•	Total Population: Has a moderate spread with a median around 9%.

•	White: Similar to the total population but with a slightly lower median.

•	African-American: Shows a higher spread and median around 14%, with some outliers.

•	Asian: Lower median around 8%, with a narrower spread.

•	Age Groups: Diabetes percentages increase with age, peaking in the 65+ years group.

•	Income Levels: Higher income correlates with lower diabetes percentages. Less than $15,000 has the highest median.

•	Education Levels: Less than high school shows higher diabetes percentages compared to college graduates.

•	Gender: Males have a slightly higher median than females.



```python

# Violin plot of Percent by Strata Name
plt.figure(figsize=(14, 8))
sns.violinplot(x='Strata Name', y='Percent', data=diabetes_data)
plt.title('Violin plot of Percent by Strata Name')
plt.xticks(rotation=90)
plt.show()
```

Observation:

•	Total Population and Whites: Have a consistent distribution around 9-10%.

•	African-Americans: Show a wider distribution, with higher variability and a median around 14%.

•	Asians: Lower median and narrow distribution around 8%.

•	Age Groups: Diabetes prevalence increases with age, peaking in the 65+ years group.

•	Income Levels: Higher income levels correlate with lower diabetes percentages, with less than $15,000 having the highest variability.

•	Education Levels: Less than high school shows higher and more variable diabetes percentages compared to college graduates.

•	Gender: Males and females have similar distributions, with males slightly higher.



```python
# Regression plot of Year vs Percent
plt.figure(figsize=(10, 6))
sns.regplot(x='Year', y='Percent', data=diabetes_data)
plt.title('Regression plot of Year vs Percent')
plt.xlabel('Year')
plt.ylabel('Percent')
plt.show()
```

Observation:

•	The regression plot shows no significant trend in the diabetes percentages from 2012 to 2017.

•	The diabetes percentages are consistently centered around 10% throughout the years.

•	There is considerable spread in the data points for each year, indicating variability in diabetes percentages within each year.

•	The confidence interval (shaded area) around the regression line is relatively narrow, suggesting a stable trend with little fluctuation over the years.

•	Some outliers are observed, especially in the higher range of diabetes percentages (above 20%) for certain years.



```python

# Pair plot of numeric columns
sns.pairplot(diabetes_data[numeric_columns])
plt.show()
```

Observation:

•	Year: No significant trends observed across years.

•	Percent: Shows a normal distribution centered around 10%.

•	Lower and Upper 95% CL: Highly correlated with the percent values, indicating consistent intervals around reported percentages.

•	Standard Error: Lower standard errors are more frequent, indicating reliable estimates.

•	Scatter Plots: Strong positive correlation between percent and both lower and upper 95% CL.

•	Histograms: Percent values show a central peak, standard error shows a peak at lower values.

•	Overall: Data indicates consistent reporting with low variability in standard errors and confidence intervals.



```python
# Correlation heatmap
plt.figure(figsize=(10, 6))
corr_matrix = diabetes_data[numeric_columns].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

Observation:

•	Year: No significant correlation with other variables.

•	Percent: Strong positive correlation with lower (0.97) and upper (0.98) 95% CL.

•	Lower 95% CL: Highly correlated with percent (0.97) and upper 95% CL (0.93).

•	Upper 95% CL: Highly correlated with percent (0.98) and lower 95% CL (0.93).

•	Standard Error: Moderate positive correlation with percent (0.4) and upper 95% CL (0.54).

•	Overall: Percent, lower 95% CL, and upper 95% CL are closely related, indicating consistency in the reported intervals around percent values.
